```python
import torch
import cv2

# Load the pre-trained YOLOv5 model (latest version)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
def detect_objects_image(image_path):
    # Load the image
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    results = model(image_rgb)
    results.show()
    results.save()
    df = results.pandas().xywh[0]
    print(df)
detect_objects_image('/content/ani.jpg')
```
