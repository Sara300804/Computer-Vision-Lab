```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import ipywidgets as widgets
from IPython.display import display
from google.colab import files

# Load YOLO model
model = YOLO('yolov8n.pt')

# Upload and extract frames
video_path = next(iter(files.upload()))
cap = cv2.VideoCapture(video_path)
frames = [frame for ret, frame in iter(lambda: cap.read(), (False, None))]
cap.release()

# Detect and display objects
def show_frame(idx):
    frame = frames[idx]
    result = model(frame)[0].plot()
    with output:
        output.clear_output(wait=True)
        plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

# Widgets
frame_slider = widgets.IntSlider(0, 0, len(frames)-1, 1, description='Frame:')

output = widgets.Output()
frame_slider.observe(lambda change: show_frame(change.new), names='value')

# Display
display(frame_slider, output)
show_frame(0)
```
